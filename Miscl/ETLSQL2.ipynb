```python
# Import and establish Base for which classes will be constructed 
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()
```

```python
# Import modules to declare columns and column data types
from sqlalchemy import Column, Integer, String, Float
```

```python
# .  does connect
engine = create_engine('postgresql://postgres:Roshan2017@localhost:5432/SPY',pool_size=20,max_overflow=-1)
#connection = engine.connect()
conn = engine.connect()
# https://martin-thoma.com/configuration-files-in-python/
```

```python
#----------------------------------------------------------------
# Create SPY class(es), with its corresponding Table
#----------------------------------------------------------------

class SPY(Base):
    __tablename__ = 'Stock_Option'
    Contract_Name = Column(String(20))
    Bid = Column(Integer, primary_key=True)
    Stocks = Column(Integer)
    Strike = Column(Integer)
    Options = Column(Integer)
    Date=Column(Integer)
    ImpliedVolatility=Column(Integer)
    ClosePrice=Column(Integer)
```
